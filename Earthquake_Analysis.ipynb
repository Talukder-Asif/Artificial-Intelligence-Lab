**DATA DISTRIBUTION**

Data distribution by Magnitudes


```python
import pandas as pd
import plotly.express as px

# Create a histogram to show the distribution of earthquake magnitudes
fig = px.histogram(
    df,
    x='magnitude',
    nbins=30,  # Adjust the number of bins as necessary
    title='Distribution of Earthquake Magnitudes',
    labels={'magnitude': 'Magnitude'},  # X-axis label
    color_discrete_sequence=['#2ca02c']  # Use a single color for all bars
)

# Update layout for better visibility
fig.update_layout(
    xaxis_title='Magnitude',  # X-axis title
    yaxis_title='Count',  # Y-axis title
    showlegend=False  # Hide legend since it’s not necessary here
)

# Show the plot
fig.show()
```

**Data distribution by Tsunami or Not**

```python
# Ensure the tsunami column exists and categorize data
if 'tsunami' in df.columns:
    # Count occurrences of tsunami (1) and non-tsunami (0)
    tsunami_counts = df['tsunami'].value_counts().reset_index()
    tsunami_counts.columns = ['Tsunami Occurrence', 'Count']

    # Map tsunami occurrence values to more descriptive labels
    tsunami_counts['Tsunami Occurrence'] = tsunami_counts['Tsunami Occurrence'].map({1: 'Tsunami', 0: 'Non-Tsunami'})

    # Create a pie chart
    fig = px.pie(
        tsunami_counts,
        names='Tsunami Occurrence',
        values='Count',
        title='Tsunami Occurrences in Earthquakes',
        color='Tsunami Occurrence',
        color_discrete_sequence=['#ff9999', '#66b3ff'],  # Colors for the pie chart
        labels={'Tsunami Occurrence': 'Tsunami Status'}
    )

    # Update layout for better visibility
    fig.update_traces(
        textposition='inside',
        textinfo='percent+label',  # Show percentage and label
        hovertemplate='Tsunami Status: %{label}<br>Count: %{value}<br>Percentage: %{percent:.2%}'  # Detailed hover info
    )

    # Show the plot
    fig.show()
else:
    print("No 'tsunami' column found in the dataset.")
```

**Data distribution by Country**

```python
# Ensure the country column exists
if 'country' in df.columns:
    # Group by country and count occurrences
    country_counts = df['country'].value_counts().reset_index()
    country_counts.columns = ['Country', 'Count']

    # Create a bar chart
    fig = px.bar(
        country_counts,
        x='Country',
        y='Count',
        title='Distribution of Earthquakes by Country',
        labels={'Country': 'Country', 'Count': 'Number of Earthquakes'},
        color='Count',  # Color bars by the count of earthquakes
        color_continuous_scale=px.colors.sequential.Blugrn  # Color scale from red to green
    )

    # Update layout for better visibility
    fig.update_layout(
        xaxis_title='Country',
        yaxis_title='Number of Earthquakes',
        xaxis_tickangle=-45,  # Rotate x-axis labels for better visibility
        showlegend=False  # Hide legend since it's not necessary here
    )

    # Show the plot
    fig.show()
else:
    print("No 'country' column found in the dataset.")
```

**Data distribution by Years**

```python
# Ensure the date column exists and convert it to datetime
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])  # Convert to datetime
    df['year'] = df['date'].dt.year  # Extract year from the date

    # Group by year and count occurrences
    year_counts = df['year'].value_counts().reset_index()
    year_counts.columns = ['Year', 'Count']

    # Sort years in ascending order
    year_counts = year_counts.sort_values('Year')

    # Create a bar chart
    fig = px.bar(
        year_counts,
        x='Year',
        y='Count',
        title='Distribution of Earthquakes by Year',
        labels={'Year': 'Year', 'Count': 'Number of Earthquakes'},
        color='Count',  # Color bars by the count of earthquakes
        color_continuous_scale=px.colors.sequential.Blugrn  # Color scale from red to green
    )

    # Update layout for better visibility
    fig.update_layout(
        xaxis_title='Year',
        yaxis_title='Number of Earthquakes',
        showlegend=False  # Hide legend since it's not necessary here
    )

    # Show the plot
    fig.show()
else:
    print("No 'date' column found in the dataset.")
```

**MAP SIMULATION**

Earthquake Map Simulation Earthquake Based on Time. In the map the color of the point changes based on the amplitude of the earthquake. Earthquakes with small amplitudes are yellow, while earthquakes with large amplitudes are dark red. Press the play button to see the simulation, and the stop button to stop it. To see the location and magnitude details, point the cursor at the earthquake point on the map.

```python
# Pastikan kolom 'date' diformat ke tipe datetime jika belum
df['date'] = pd.to_datetime(df['date'])

# Tentukan rentang warna untuk magnitudo
min_magnitude = df['magnitude'].min()
max_magnitude = df['magnitude'].max()

# Plot animasi menggunakan scatter_mapbox dari Plotly Express
fig = px.scatter_mapbox(
    df,
    lat="latitude",
    lon="longitude",
    size="magnitude",  # Kolom magnitude digunakan sebagai ukuran lingkaran
    color="magnitude",  # Menggunakan magnitudo untuk warna
    color_continuous_scale=px.colors.sequential.YlOrRd,  # Skala warna dari hijau ke merah
    range_color=[min_magnitude, max_magnitude],  # Mengatur rentang magnitudo
    animation_frame="date",  # Animasi berdasarkan waktu
    zoom=0.5,  # Zoom yang sesuai untuk peta global
    height=600,
    title="Earthquake Magnitude Visualization",
    text="location"  # Menambahkan label lokasi di atas titik
)

# Menambahkan gaya tampilan teks dan transparansi marker
fig.update_traces(
    textposition="top center",  # Posisi label di atas titik
    textfont=dict(color="black"),  # Warna teks menjadi hitam
    marker=dict(opacity=0.7)  # Mengatur opasitas lingkaran agar lebih transparan
)

# Mengatur gaya peta menjadi open-street-map
fig.update_layout(mapbox_style="open-street-map")

# Menampilkan peta
fig.show()
```
